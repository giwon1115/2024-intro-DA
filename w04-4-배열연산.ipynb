#배열과 배열의 연산

```python
a=np.arange(1,7).reshape(2,3)
a
```

```python
b=np.full_like(a,[1,2,3])
b
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
a=np.arange(4)
a
```

```python
b=np.arange(0.5,4,1)
b
```

```python
a+=b
```

```python
b+=a
b
```

```python
a=np.full((2,3),[1,2,3])
a
```

```python
b=np.full((3,2),[2,1])
b
```

```python
a@b
```

```python
a.dot(b)
```

```python
b.dot(a)
```

#배열 함수와 범용함수

배열함수

```python
a=np.array([[10,20],[5,8],[1,2]])
a
```

```python
a.sum()
```

```python
a.sum(axis=0)
```

```python
a.sum(axis=1)
```

```python
print(np.sum(a))
print(np.sum(a,axis=0))
print(np.sum(a,axis=1))
```

```python
a=np.array([[3,5,8],[1,2,3]])
a
```

```python
a.cumsum()
```

```python
a.cumsum(axis=0)
```

```python
a.cumsum(axis=1)
```
